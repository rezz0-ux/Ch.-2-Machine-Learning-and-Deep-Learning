### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import cluster
from sklearn.metrics import silhouette_score
```

### Read Dataset

```python
data = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/cluster_s1.csv')
data = data.drop('no', axis = 1)
data.head()
```

```python
%matplotlib inline
ax1 = data.plot.scatter(x='x', y='y', c='DarkBlue')
```

### Data Segmentation (Clustering)

```python
# TODO: Lakukan pencarian jumlah cluster terbaik berdasarkan data diatas
silhouette_scores = []
K_range = range(2, 21)

for k in K_range:
    kmeans = cluster.KMeans(n_clusters=k, random_state=0)
    kmeans.fit(data)
    silhouette_avg = silhouette_score(data, kmeans.labels_)
    print(f"For n_clusters={k}, the average silhouette_score is: {silhouette_avg:.2f}")
```

```python
# TODO: Lakukan pemodelan dengan KMeans
kmeans = cluster.KMeans(n_clusters=15, random_state=0)
kmeans.fit(data)
label = kmeans.labels_

score = silhouette_score(data, label)
print("silhouette score={:.5f}".format(score))
```

### Visualization: Cluster Result

```python
# TODO: Masukkan label hasil clustering ke dalam DataFrame

result = data.copy()
result['cluster'] = label
result.head()
```

```python
# TODO: Plot data hasil clustering dengan Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x='x', y='y', hue='cluster', data=result, palette='tab20', legend='full')
plt.title(f'Clustering Result with k=15')
plt.show()
```
